```python
from sympy import Symbol, symbols, sin, cos, diff, solveset, S, pprint
from sympy.plotting import plot as symplt
from art import tprint
```

```python
x = Symbol('x')
func = symplt(sin(x)**2 - cos(x)**2, show=False, line_color = 'indigo')
func.show()
```

```python
# производная + график
eq_diff = diff(sin(x)**2 - cos(x)**2, x)
symplt(eq_diff, show=False, line_color = 'red').show()
eq_diff
```

```python
# нули функции
pprint(solveset(sin(x)**2 - cos(x)**2, x, domain=S.Reals), use_unicode=True)
```

```python
# больше и меньше 0
pprint(solveset(sin(x)**2 - cos(x)**2 < 0, x, domain=S.Reals), use_unicode=True)
```

```python
pprint(solveset(sin(x)**2 - cos(x)**2 > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# возрастание и убывание
pprint(solveset(eq_diff > 0, x, domain=S.Reals), use_unicode=True)
```

```python
pprint(solveset(eq_diff < 0, x, domain=S.Reals), use_unicode=True)
```
